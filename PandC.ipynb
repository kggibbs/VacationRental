```python
import datetime as dt
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas.plotting import register_matplotlib_converters
import seaborn as sn

register_matplotlib_converters()
%matplotlib inline
pd.set_option('precision', 3)
pd.options.display.float_format = '{:.2f}'.format
```

```python
df1 = pd.read_csv('/home/gibbs/SJIs/VacationRentals/Data/Parcels_and_Condominiums.csv',
                  dtype={'Use_Code':np.int32,'ALTPin':object},
                  header=0,nrows=18803,low_memory=False)
```

```python
def isSFRQ(z):
    Use_Code, Descriptio = z
    if ('IMPROV ONLY' in Descriptio) or ('IMPROVEMENT ONLY' in Descriptio):
        return False
    if (Use_Code < 1200) or ((Use_Code >= 1900) and (Use_Code <= 1999)):
        return True
    else:
        return False

def isOnORCASQ(z):
    Tax_Area, Short_Lega, District_N = z
    if 'ORCAS' not in Tax_Area:
        return False
    if ('OBSTRUCTION' in Short_Lega) or ('CRANE' in Short_Lega):
        return False
    if 'Other' in District_N:
        return False
    return True

df1['isSFR'] = [isSFRQ(z) for z in zip(df1['Use_Code'],df1['Descriptio'])]

df1['inOnORCAS'] = [isOnORCASQ(z) for z in zip(df1['Tax_Area'],df1['Short_Lega'],df1['District_N'])]
```

```python
dfALLsfr = df1[(df1['isSFR']) & (df1['Bldg_Value'] >= 42000)]

dfORCASsfr = df1[(df1['inOnORCAS']) & (df1['isSFR']) & (df1['Bldg_Value'] >= 42000)]

dfLOPEZsfr = df1[(df1['Tax_Area'].str.contains('LOPEZ',case=False)) \
    & (df1['isSFR']) & (df1['Bldg_Value'] >= 42000)]

dfSJIALLsfr = df1[((df1['Tax_Area'].str.contains('JUAN',case=False)) \
    | (df1['Tax_Area'].str.contains('FRIDAY',case=False))) \
    & (df1['isSFR']) & (df1['Bldg_Value'] >= 42000)]

dfSJIsfr = df1[(df1['Tax_Area'].str.contains('JUAN',case=False)) \
    & (df1['isSFR']) & (df1['Bldg_Value'] >= 42000)]
```
