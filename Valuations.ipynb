```python
import datetime as dt
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas.plotting import register_matplotlib_converters
import seaborn as sn

register_matplotlib_converters()
%matplotlib inline
pd.set_option('precision', 3)
pd.options.display.float_format = '{:.2f}'.format
```

```python
df1 = pd.read_csv('/home/gibbs/SJIs/VacationRentals/Data/results-2018-19_20190515.txt',header=0)

df1['val'].describe()
```

```python
nlow = 0
nmed = 0
nhigh = 0
nvhigh = 0
for val in df1['val']:
    if val <= 300000:
        nlow += 1
    elif val > 300000 and val <= 400000:
        nmed += 1
    elif val > 40000 and val <= 1600000:
        nhigh += 1
    else:
        nvhigh += 1
total = nlow+nmed+nhigh+nvhigh
(nlow, nmed, nhigh, nvhigh, total) 
```

```python
(round(100*nlow/total,1),round(100*nmed/total,1),
 round(100*nhigh/total,1),round(100*nvhigh/total,1),
 round(100*(nlow+nmed)/total,1))
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(11.7, 8.27)

xbins = list(range(0,4800,25))
xtixs = np.arange(0,4800,600)
ax.set_xticks(xtixs)

plt.rcParams["patch.force_edgecolor"] = True
plt.title('Distribution of Assessed Values\nfor Permitted Vacation Rentals',fontweight='bold', fontsize='17')

#plt.hist(df['val']/1000,bins=40)
N, bins, patches = plt.hist(df1['val']/1000,xbins)

#cmap = plt.get_cmap('terrain')
#clow = cmap(0.3)
#cmedium = cmap(0.5)
#chigh = cmap(0.9)

#cmap = plt.get_cmap('jet')
#low = cmap(0.5)
#medium = cmap(0.3)
#high = cmap(0.9)

for i in range(0,12):
    patches[i].set_facecolor('slateblue')
for i in range(12,16):
    patches[i].set_facecolor('darkslateblue')
for i in range(16,191):
    patches[i].set_facecolor('lightslategray')
#for i in range(11,30):
#    patches[i].set_facecolor(chigh)
plt.rcParams['ytick.labelsize'] = 20
plt.rcParams['xtick.labelsize'] = 20

plt.xlabel('Assessed Value (1000s $)', fontweight='bold', fontsize='17', horizontalalignment='center', labelpad=20)

# from matplotlib.patches import Ellipse
# el = Ellipse((2, -1), 0.5, 0.5)
# ax.add_patch(el)
# ax.annotate('12%',
#             xy=(200, 20), xycoords='data',
#             xytext=(-20, 25), textcoords='data',
#             size=20,
#             # bbox=dict(boxstyle="round", fc="0.8"),
#             arrowprops=dict(arrowstyle="fancy",
#                             fc=clow, ec="black",
#                             patchB=el,
#                             connectionstyle="angle3,angleA=0,angleB=90"))


# #el = Ellipse((2, -1), 0.5, 0.5)
# #ax.add_patch(el)
# ax.annotate('15%',
#             xy=(300, 35), xycoords='data',
#             xytext=(0, 40), textcoords='data',
#             size=20,
#             # bbox=dict(boxstyle="round", fc="0.8"),
#             arrowprops=dict(arrowstyle="fancy",
#                             fc=cmedium, ec="black",
#                             patchB=el,
#                             connectionstyle="angle3,angleA=0,angleB=110"))

fig.savefig('Assessed-2018-2019.png')
```
